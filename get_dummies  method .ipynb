```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_excel('one_hot.xlsx')
df
```

```python
dummies = pd.get_dummies(df.city)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns')
merged
```

```python
final = merged.drop(['city','Hyd'],axis ='columns')
final
```

```python
reg = linear_model.LinearRegression()
reg.fit(final[['area','Banglore','Pune']],final.price)
```

```python
reg.predict([[2800,1,0]])
```

```python
reg.score(final[['area','Banglore','Pune']],final.price)
```
