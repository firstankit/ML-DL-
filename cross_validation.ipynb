```python
import pandas as pd
```

```python
df = pd.read_csv('Train.csv')
df
```

```python
df.describe()
```

```python
df['Age']= df['Age'].fillna(30)
```

```python
df.describe()
```

```python
x = df.drop(['PassengerId','Survived','Name','Ticket','Cabin'],axis = 1)
x.head()
```

```python
y = df.Survived
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = x
dfle['Sex'] = le.fit_transform(dfle['Sex'])
dfle
```

```python
dfle['Embarked'] = le.fit_transform(dfle['Embarked'])
dfle
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dfle,y,test_size=0.3)
```

```python
lr = LogisticRegression(solver='liblinear',multi_class='ovr')
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
svm = SVC(gamma='auto')
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lrc=cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), dfle,y,cv=10)
```

```python
svc=cross_val_score(SVC(gamma='auto'), dfle,y,cv=10)
```

```python
rfv=cross_val_score(RandomForestClassifier(n_estimators=10),dfle,y,cv=10)
```

```python
np.mean(lrc)
```

```python
np.mean(svc)
```
