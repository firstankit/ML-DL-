# Importing  Libraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Sample_for_AI.csv')
df
```

# Data Preprocessing 

```python
df.isnull().sum()
```

```python
df['Origin'].unique()
```

```python
df['Specialization'].unique()
```

```python
df1 = df.drop(['Specialization'],axis =1)
# Specialization have 3380 Null values out of 9240 sample so removing it
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

# Removing null Values

```python
df2=df1.dropna()
```

```python
df2.isnull().sum()
```

# Converting test to no. using one hot encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dfle = df2
dfle.Purchased = le.fit_transform(dfle.Purchased)
dfle
```

```python
dfle['Do Not Email'] = le.fit_transform(dfle['Do Not Email'])
dfle
```

```python
dfle['Do Not Call'] = le.fit_transform(dfle['Do Not Call'])
dfle
```

```python
dfle['Origin'] = le.fit_transform(dfle['Origin'])
dfle
```

```python
dfle.describe()
```

```python
dfle.isnull().sum()
```

# Seprating Dependent and Independent Variables

```python
y = dfle.Purchased
```

```python
y.head()
```

```python
x = dfle.drop(['Purchased','ID'],axis = 1)
```

```python
x.head()
```

# Converting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

# Using SVM MODEL

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Using Random Forest MODEL

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

# Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

# Using Decision Tree

```python
from sklearn import tree
model3 = tree.DecisionTreeClassifier()
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

# Using Naive Bayes

```python
from sklearn .naive_bayes import GaussianNB
model4 = GaussianNB()
model4.fit(x_train,y_train)
```

```python
model4.score(x_test,y_test)
```

# LOGISTIC REGRESSION GIVES THE BEST RESULT 77.37% ACCURACY
