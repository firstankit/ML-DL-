```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Titanic.csv')
df.head()
```

```python
df.groupby('Survived')['Survived'].agg('count')
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis = 'columns',inplace = True)
df.head()
```

```python
inputs = df.drop(['Survived'],axis = 'columns')
```

```python
Target = df.Survived
```

```python
inputs.head()
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
df_n = pd.concat([inputs,dummies],axis = 'columns')
df_n
```

```python
df_new = df_n.drop(['Sex','female'],axis = 'columns')
```

```python
df_new.describe()
```

```python
df_new.fillna({
    'Age':29.69
},inplace = True)
```

```python
df_new.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_new,Target,test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[2,60,5,0]])
```
